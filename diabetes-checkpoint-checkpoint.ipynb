```python
diabetes_data_list=[1]
```

```python
with_diabetes=diabetes[diabetes.Outcome.isin(diabetes_data_list)]
```

```python
sns.heatmap(with_diabetes.corr(), annot = True)
```

```python
plt.figure(figsize=(15,6))
sns.countplot(x ="Pregnancies", data = with_diabetes)
plt.title("Number of pregnancies with diabetes", fontsize = 20);
```

```python
plt.figure(figsize=(27,10))
glucose=sns.countplot(x ="Glucose", data = with_diabetes)
glucose.set_xticklabels(glucose.get_xticklabels(), rotation = 90, fontsize=16)
plt.title("Amount of Glucose in a diabetic pregnant patient", fontsize = 20);
```

```python
plt.figure(figsize=(15,6))
sns.countplot(x ="BloodPressure", data = with_diabetes)
plt.title("Bloodpressure of a diabetic pregnant patient", fontsize = 20);
```

```python
with_diabetes.BloodPressure.unique()
with_diabetes_data_list=[70]
Bloodpressure=with_diabetes[with_diabetes.BloodPressure.isin(with_diabetes_data_list)]
Bloodpressure
```

```python
plt.figure(figsize=(15,6))
sns.countplot(x ="SkinThickness", data = with_diabetes)
plt.title("Skinthickness of people with diabetes", fontsize = 20);
```

```python
with_diabetes.SkinThickness.unique()
with_diabetes_data_list=[0]
Skinthickness=with_diabetes[with_diabetes.SkinThickness.isin(with_diabetes_data_list)]
Skinthickness
```

```python
plt.figure(figsize=(15,6))
insulin=sns.countplot(x ="Insulin", data = with_diabetes)
insulin.set_xticklabels(insulin.get_xticklabels(), rotation = 90, fontsize=10)
plt.title("The number of Insulin present in the of those with diabetes", fontsize = 20);
```

```python
with_diabetes.Insulin.unique()
with_diabetes_data_list=[0]
insulin=with_diabetes[with_diabetes.Insulin.isin(with_diabetes_data_list)]
insulin
```

```python
plt.figure(figsize=(23,6))
bmi=sns.countplot(x ="BMI", data = with_diabetes)
bmi.set_xticklabels(bmi.get_xticklabels(), rotation = 90, fontsize=10)
plt.title("BMI of people with diabetes", fontsize = 20);
```

```python
plt.figure(figsize=(23,6))
dpf=sns.countplot(x ="DiabetesPedigreeFunction", data = with_diabetes)
dpf.set_xticklabels(dpf.get_xticklabels(), rotation = 90, fontsize=10)
plt.title("DiabetesPredigreeFunctin", fontsize = 20);
```

```python
plt.figure(figsize=(23,6))
sns.countplot(x ="Age", data = with_diabetes)
plt.title("Age", fontsize = 20);
```
